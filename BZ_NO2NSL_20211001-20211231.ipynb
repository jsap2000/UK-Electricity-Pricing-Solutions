```python
# Python code to find average of each consecutive segment
  
# Importing
import statistics
from itertools import islice
import numpy as np

# Change CET-CEST 27/28-Mar-21 i.e. row 2063-2064
# Change CEST-CET 30/31-Oct-21 i.e. row 7271-7272

# 01-Oct-21 00:00-01:00 is in column 6552

# Days til CET-CEST change from 01-Jan-21
dtcest = int(len(BZNO2NSLDA2021.iloc[0:(2063+1),0])/24)
print(dtcest)

#Days til CEST-CET change from 01-Jan-21
dtcst = int(len(BZNO2NSLDA2021.iloc[0:(7271+1),0])/24)
print(dtcst)

def function():
    result = []
    for i in range(0, len(BZNO2NSLDA2021.iloc[:,0])):
        if i < dtst:
            number = i
            start = 24*i
            end = 24*(i+1)-1+1
            day = BZNO2NSLDA2021.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2021.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif i == 86: # For some reason this day does not show in the DataFrame
            number = i
            start = 24*i
            end = 2087+1
            day = BZNO2NSLDA2021.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2021.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif 87 <= i < dtcst:
            number = i
            start = 24*i
            end = 24*(i+1)-1+1
            day = BZNO2NSLDA2021.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2021.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif i == 303:
            number = i
            start = 24*i
            end = 7296+1
            day = BZNO2NSLDA2021.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2021.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif 304 <= i < 365:
            number = i
            start = 24*i+1
            end = 24*(i+1)-1+2
            day = BZNO2NSLDA2021.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2021.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})   
    df = pd.DataFrame(result, columns=['Day','Daily DA'])
    return df
print(function())
```
