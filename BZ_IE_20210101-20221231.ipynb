```python
import pandas as pd
import numpy as np

%store -r BZIEDA2021Daily
BZIEDA2021Daily
```

```python
%store -r BZIEDA2022Daily
BZIEDA2022Daily
```

```python
BZIE = pd.concat([BZIEDA2021Daily["Daily DA"], BZIEDA2022Daily["Daily DA"]], ignore_index=True).to_frame("BZN|IE")

BZIE["Date"] = pd.date_range(start="1/1/2021", periods=730, freq="D")

NewOrder = ["Date", "BZN|IE"]

BZIE = BZIE.reindex(columns=NewOrder)
BZIE
```

```python
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

plt.title('IE(SEM) Electricity Day-Ahead Price')
plt.ylabel('Day-Ahead Electricty Price [EUR/MWh]')

ax.plot(BZIE["Date"], BZIE["BZN|IE"], label="IE DA")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

fig.autofmt_xdate()

print(ax.get_xlim())

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
#plt.xlim(18591.55, 19379.8)
plt.tight_layout()

plt.show()
```
