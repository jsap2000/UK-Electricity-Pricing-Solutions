```python
# Python code to find average of each consecutive segment
  
# Importing
import statistics
from itertools import islice
import numpy as np

def function():
    result = []
    for i in range(0, len(NEM2021.iloc[:,0])):
        if 0 <= i < 86:
            start = i*96
            end = 96 + i*96
            day = NEM2021.iloc[start,0]
            operation1 = sum(NEM2021.iloc[start:end,1])/4
            operation2 = sum(NEM2021.iloc[start:end,2])/4
            result.append({"Day": day, "BZN|BE > BZN|GB [MW]": operation1, "BZN|GB > BZN|BE [MW]": operation2})
        elif i == 86:
            start = i*96
            skip = 8264
            end = 8352
            day = NEM2021.iloc[start,0]
            operation1 = (sum(NEM2021.iloc[start:skip,1]) + sum(NEM2021.iloc[(skip+4):end,1]))/4
            operation2 = (sum(NEM2021.iloc[start:skip,2]) + sum(NEM2021.iloc[(skip+4):end,2]))/4
            result.append({"Day": day, "BZN|BE > BZN|GB [MW]": operation1, "BZN|GB > BZN|BE [MW]": operation2})
        elif 87 <= i < 303:
            start = i*96
            end = 96 + i*96
            day = NEM2021.iloc[start,0]
            operation1 = sum(NEM2021.iloc[start:end,1])/4
            operation2 = sum(NEM2021.iloc[start:end,2])/4
            result.append({"Day": day, "BZN|BE > BZN|GB [MW]": operation1, "BZN|GB > BZN|BE [MW]": operation2})
        elif i == 303:
            start = i*96
            end = 29188
            day = NEM2021.iloc[start,0]
            operation1 = sum(NEM2021.iloc[start:end,1])/4
            operation2 = sum(NEM2021.iloc[start:end,2])/4
            result.append({"Day": day, "BZN|BE > BZN|GB [MW]": operation1, "BZN|GB > BZN|BE [MW]": operation2})
        elif 304 <= i < 365:
            start = i*96 + 4
            end = 100 + i*96
            day = NEM2021.iloc[start,0]
            operation1 = sum(NEM2021.iloc[start:end,1])/4
            operation2 = sum(NEM2021.iloc[start:end,2])/4
            result.append({"Day": day, "BZN|BE > BZN|GB [MW]": operation1, "BZN|GB > BZN|BE [MW]": operation2})   
    df = pd.DataFrame(result, columns=['Day','BZN|BE > BZN|GB [MW]', 'BZN|GB > BZN|BE [MW]'])
    return df
print(function())
```
