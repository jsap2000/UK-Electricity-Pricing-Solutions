```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

LNGGB = pd.read_csv(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\LNG_GB_20210101-20221231.csv")

LNGSt2022 = LNGGB.iloc[:365,:]
LNGSt2022 = LNGSt2022.reset_index()
del LNGSt2022['index']

LNGSt2021 = LNGGB.iloc[365:,:]
LNGSt2021 = LNGSt2021.reset_index()
del LNGSt2021['index']

print(LNGSt2021.to_string())
```

```python
GGB2021 = pd.read_csv(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\GAS_GB_20210101-20211231.csv")

LRSt2021 = GGB2021.loc[GGB2021["Data Item"] == "Storage, Long Range, Stock Levels"]
LRSt2021 = LRSt2021.reset_index()
del LRSt2021['index']

MRSt2021 = GGB2021.loc[GGB2021["Data Item"] == "Storage, Medium Range, Stock Levels"]
MRSt2021 = MRSt2021.reset_index()
del MRSt2021['index']

SRSt2021 = GGB2021.loc[GGB2021["Data Item"] == "Storage, Short Range, Stock Levels"]
SRSt2021 = SRSt2021.reset_index()
del SRSt2021['index']

GStGB2021 = pd.DataFrame({"Date": LRSt2021.iloc[:,1],
    "Long Range [GWh]": LRSt2021.iloc[:,3]})
GStGB2021["Medium Range [GWh]"] = MRSt2021.iloc[:,3]
GStGB2021["Short Range [GWh]"] = SRSt2021.iloc[:,3]
GStGB2021["LNG [GWh]"] = LNGSt2021.iloc[:,3]

GStGB2021 = GStGB2021.iloc[::-1]
GStGB2021 = GStGB2021.reset_index()
del GStGB2021['index']

print(GStGB2021.to_string())
```

```python
GGB2022 = pd.read_csv(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\GAS_GB_20220101-20221231.csv")

LRSt2022 = GGB2022.loc[GGB2022["Data Item"] == "Storage, Long Range, Stock Levels"]
LRSt2022 = LRSt2022.reset_index()
del LRSt2022['index']

MRSt2022 = GGB2022.loc[GGB2022["Data Item"] == "Storage, Medium Range, Stock Levels"]
MRSt2022 = MRSt2022.reset_index()
del MRSt2022['index']

SRSt2022 = GGB2022.loc[GGB2022["Data Item"] == "Storage, Short Range, Stock Levels"]
SRSt2022 = SRSt2022.reset_index()
del SRSt2022['index']

GStGB2022 = pd.DataFrame({"Date": LRSt2022.iloc[:,1],
    "Long Range [GWh]": LRSt2022.iloc[:,3]})
GStGB2022["Medium Range [GWh]"] = MRSt2022.iloc[:,3]
GStGB2022["Short Range [GWh]"] = SRSt2022.iloc[:,3]
GStGB2022["LNG [GWh]"] = LNGSt2022.iloc[:,3]

GStGB2022 = GStGB2022.iloc[::-1]
GStGB2022 = GStGB2022.reset_index()
del GStGB2022['index']

print(GStGB2022.to_string())
```

```python
GStGB = pd.DataFrame({"Date": pd.date_range(start="1/1/2021", periods=730, freq="D")})
GStGB["Long Range [GWh]"] = pd.concat([GStGB2021["Long Range [GWh]"], GStGB2022["Long Range [GWh]"]], ignore_index=True)
GStGB["Medium Range [GWh]"] = pd.concat([GStGB2021["Medium Range [GWh]"], GStGB2022["Medium Range [GWh]"]], ignore_index=True)
GStGB["Short Range [GWh]"] = pd.concat([GStGB2021["Short Range [GWh]"], GStGB2022["Short Range [GWh]"]], ignore_index=True)
GStGB["LNG [GWh]"] = pd.concat([GStGB2021["LNG [GWh]"], GStGB2022["LNG [GWh]"]], ignore_index=True)

GStGB.iloc[:,1:] = GStGB.iloc[:,1:].round(1)

print(GStGB.to_string())
```

```python
NewDF = pd.DataFrame()
NewDF["Date"] = GStGB["Date"]
NewDF["Storage"] = GStGB["Long Range [GWh]"] + GStGB["Medium Range [GWh]"] + GStGB["Short Range [GWh]"] + GStGB["LNG [GWh]"]

print(NewDF.to_string())

print(min(NewDF["Storage"]))
```

```python
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
import matplotlib.ticker as ticker

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

plt.title('Gas Storage Changes in the UK')
plt.ylabel('Compound Quantity of Natural Gas Injected or Withdrawn from Storage [?]')

Mean = GasStorGB2["Storage Injection [?]"] + GasStorGB2["Storage Withdrawal [?]"]

ax.plot(GasStorGB2["Date"], GasStorGB2["Storage Injection [?]"], color = "darkgreen", label="Storage Injections")
ax.plot(GasStorGB2["Date"], GasStorGB2["Storage Withdrawal [?]"], color = "firebrick", label="Storage Withdrawal")

plt.fill_between(GasStorGB2["Date"], GasStorGB2["Storage Injection [?]"], color="darkgreen")
plt.fill_between(GasStorGB2["Date"], GasStorGB2["Storage Withdrawal [?]"], color="firebrick")

ax.plot(GasStorGB2["Date"], Mean.rolling(28,  center=True).mean(), linewidth=3, color = "gold", label="Rolling Average")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.xlim(18628, 19358)
plt.ylim(-1e9, 1e9)
plt.tight_layout()

ax.xaxis.set_minor_locator(mdates.MonthLocator())
ax.xaxis.set_minor_formatter(mdates.DateFormatter(""))

plt.show()
```
