```python
dFEMix = EMix.loc[EMix["Continent"] == "Europe"].loc[EMix["Year"] == 2021].loc[EMix["Category"] == "Electricity demand"].loc[EMix["Subcategory"] == "Demand"].loc[EMix["Unit"] == "TWh"]
dFEMix = dFEMix.reset_index()
dFEMix = dFEMix.drop([14], axis=0)
dFEMix = dFEMix.reset_index()
del dFEMix['index']
del dFEMix['level_0']

print(dFEMix.to_string())

fFEMix = EMix.loc[EMix["Continent"] == "Europe"].loc[EMix["Year"] == 2021].loc[EMix["Category"] == "Electricity imports"]
fFEMix = fFEMix.reset_index()
fFEMix = fFEMix.drop([14], axis=0)
fFEMix = fFEMix.reset_index()
del fFEMix['index']
del fFEMix['level_0']

print(fFEMix.to_string())

gFEMix = EMix.loc[EMix["Continent"] == "Europe"].loc[EMix["Year"] == 2021].loc[EMix["Category"] == "Electricity generation"].loc[EMix["Subcategory"] == "Fuel"].loc[EMix["Variable"] == "Gas"].loc[EMix["Unit"] == "%"]
gFEMix = gFEMix.sort_values("Area", ascending=True)
gFEMix = gFEMix.reset_index()
del gFEMix['index']

print(gFEMix.to_string())
```

```python
eFEMix = pd.DataFrame({"Country": dFEMix.iloc[:,0],
    "Annual Demand [TWh]": dFEMix.iloc[:,14]})
eFEMix["Gas Generation [%]"] = gFEMix.iloc[:,14]
eFEMix["Annual Imports [TWh]"] = fFEMix.iloc[:,14]
eFEMix["Import Dependance [%]"] = round(fFEMix.iloc[:,14]/dFEMix.iloc[:,14]*100, 2)



eFEMix = eFEMix.replace(np.NaN, 0)

eFEMix = eFEMix.reset_index()
del eFEMix['index']
print(eFEMix)
```
