```python
# importing the required module
import matplotlib.pyplot as plt
import numpy as np
  
# x axis values
x = UKEGS.iloc[43,2:]

# corresponding y axis values
Nuclear = UKEGS.iloc[72,2:]
NFH = UKEGS.iloc[73,2:]
TWind = UKEGS.iloc[74,2:]
Tidal = UKEGS.iloc[77,2:]
Solar = UKEGS.iloc[78,2:]
Bioenergy = UKEGS.iloc[79,2:]
PSH = UKEGS.iloc[82,2:]

exclude_values = "[x]"
Tidal[np.isin(Tidal, exclude_values)] = np.nan
Solar[np.isin(Solar, exclude_values)] = np.nan

# plotting the points 
plt.plot(x, Nuclear, label="Nuclear")
plt.plot(x, NFH, label="NFH")
plt.plot(x, TWind, label="Wind (Total)")
plt.plot(x[~np.isnan(Tidal)], Tidal[~np.isnan(Tidal)], label="Tidal")
plt.plot(x[~[x]], Solar[~[x]], label="Solar")
plt.plot(x, Bioenergy, label="Bioenergy")
plt.plot(x, PSH, label="PSH")

plt.ylabel('Electricity Generated [TWh]')
  
plt.legend()
plt.title('GB Non-Renewable Quarterly Electricity Generation')

plt.tick_params(axis='x', which='major', labelsize=5)
plt.setp(plot.axes.get_xticklabels(), visible=False)
plt.setp(plot.axes.get_xticklabels()[::40], visible=True)

# function to show the plot
plt.show()
```

```python
import numpy as np

# Define data values by creating a Dataframe using a n-dimensional list
df = pd.DataFrame([[7, 5, 3], [14, 12, 5], [21, 19, 11],
                   [28, 21, 20], [35, 31, 15], [42, 27, 29],
                   [49, 35, 31]])

df.rename(columns={0: 'x', 1: 'y', 2: 'z'}, inplace=True)

print(np.shape(df), type(df), df, sep='\n')

import matplotlib.pyplot as plt

# Plot a simple line chart
plt.plot(df['x'], df['y'], color='g', label='Line y')

# Plot another line on the same chart/graph
plt.plot(df['x'], df['z'], color='r', label='Line z')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set some values of y to be excluded
exclude_values = [0.5, 0.7, 0.9]
y[np.isin(y, exclude_values)] = np.nan

# Plot the data while excluding specific values
plt.plot(x[~np.isnan(y)], y[~np.isnan(y)], 'o')
plt.show()
```
