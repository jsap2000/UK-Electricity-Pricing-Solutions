```python
import matplotlib.pyplot as plt

# Set x-axis
Quarter = UKEGS.iloc[43,2:]

# Set y-axis data series
Coal = UKEGS.iloc[69,2:]
Oil = UKEGS.iloc[70,2:]
Gas = UKEGS.iloc[71,2:]
Nuclear = UKEGS.iloc[72,2:]

# Create a subplot of given dimensions
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

plt.title('GB Non-Renewable Quarterly Electricity Generation')
plt.ylabel('Electricity Generated [TWh]')

# Plot individual data series one by one
ax.plot(Quarter, Coal, label="Coal")
ax.plot(Quarter, Oil, label="Oil")
ax.plot(Quarter, Gas, label="Gas")
ax.plot(Quarter, Nuclear, label="Nuclear")

plt.legend()

# Set gridline styles
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

# Set the x-axis tick interval and add minor ticks on the y-axis
n = 4
ax.xaxis.set_minor_locator(plt.MultipleLocator(n-3))
ax.xaxis.set_major_locator(plt.MultipleLocator(n))
ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

# Format the labels on the x-axis
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
# Set x-axis
x = UKEGS.iloc[43,2:]

# corresponding y axis values
#Nuclear = UKEGS.iloc[72,2:]
NFH = UKEGS.iloc[73,2:]
TWind = UKEGS.iloc[74,2:]
Tidal = UKEGS.iloc[77,2:]
Solar = UKEGS.iloc[78,2:]
Bioenergy = UKEGS.iloc[79,2:]
PSH = UKEGS.iloc[81,2:]

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

plt.title('GB Non-Renewable Quarterly Electricity Generation')
plt.ylabel('Electricity Generated [TWh]')

# Covert the Tidal and Solar data to float format to ignore [x] data
Tidal = Tidal.astype(float)
Tidal = Tidal.interpolate(method='linear')

Solar = Solar.astype(float)
Solar = Solar.interpolate(method='linear')

ax.plot(x, NFH, label="NFH")
ax.plot(x, TWind, label="Wind (Total)")
ax.plot(x, Tidal, label="Tidal")
ax.plot(x, Solar, label="Solar")
ax.plot(x, Bioenergy, label="Bioenergy")
ax.plot(x, PSH, label="PSH")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

n = 4
ax.xaxis.set_minor_locator(plt.MultipleLocator(n-3))
ax.xaxis.set_major_locator(plt.MultipleLocator(n))
ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
