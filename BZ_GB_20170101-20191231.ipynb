```python
import statistics

def function():
    result = []
    for i in range(0, len(BZGBDA2017.iloc[:,0])):
        if i < 84:
            start = 24*i
            end = 24*(i+1)
            day = BZGBDA2017.iloc[start,0]
            operation = statistics.mean(BZGBDA2017.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif i == 84: # For some reason this day does not show in the DataFrame
            start = 24*i
            skip = 2018
            end = 2040
            a = sum(BZGBDA2017.iloc[start:skip,1])
            b = sum(BZGBDA2017.iloc[skip+1:end,1])
            day = BZGBDA2017.iloc[start,0]
            operation = (a+b)/(len(BZGBDA2017.iloc[start:end,1])-1)
            result.append({"Day": day, "Daily DA": operation})
        elif 85 <= i < 301:
            start = 24*i
            end = 24*(i+1)
            day = BZGBDA2017.iloc[start,0]
            operation = statistics.mean(BZGBDA2017.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif i == 301:
            start = 24*i
            end = 7249
            day = BZGBDA2017.iloc[start,0]
            operation = statistics.mean(BZGBDA2017.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif 302 <= i < 365:
            start = 24*i+1
            end = 24*(i+1)+1
            day = BZGBDA2017.iloc[start,0]
            operation = statistics.mean(BZGBDA2017.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})   
    df = pd.DataFrame(result, columns=['Day','Daily DA'])
    return df

BZGBDA2017Daily = function()
BZGBDA2017Daily["Daily DA"] = BZGBDA2017Daily["Daily DA"].round(2)
print(BZGBDA2017Daily.to_string())
```
