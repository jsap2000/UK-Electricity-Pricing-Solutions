```python
import pandas as pd
import numpy as np
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker

# Gas data for Great Britain

%store -r rhoSAP_BZNGB
%store -r pSAP_BZNGB
%store -r CompGasGB
%store -r MSAP
%store -r CSAP

# Generation data for Great Britain

%store -r rhoGBW_BZNGB
%store -r pGBW_BZNGB
%store -r GBGen
%store -r MGBW
%store -r CGBW
```

```python
# Collect the base data that we want into a DataFrame
GBWeight = pd.DataFrame()
GBWeight["Date"] = GBGen["Date"]
GBWeight["Electricity Price [EUR/MWh]"] = GBGen["GB Electricity DA [EUR/MWh]"].astype(float)
GBWeight["Gas Price [EUR/MWh]"] = CompGasGB["GB Gas SAP [EUR/MWh]"].astype(float)
GBWeight["Wind Power [MWh]"] = GBGen["Wind [MWh]"].astype(float)
```

```python
# Set up a second DataFrame for correlation related based estimates
GBWeight2 = pd.DataFrame()
GBWeight2["Date"] = GBWeight["Date"]
GBWeight2["Electricity Price [EUR/MWh]"] = GBWeight["Electricity Price [EUR/MWh]"]

# Add the data for gas based on the correlation coefficient
GBWeight2["Gas Factor [EUR/MWh]"] = MSAP*GBWeight["Gas Price [EUR/MWh]"]+CSAP
GBWeight2["Gas Factor Unc. [EUR/MWh]"] = abs((1-rhoSAP_BZNGB)*(MSAP*GBWeight["Gas Price [EUR/MWh]"]+CSAP))
GBWeight2["Gas Factor Min. [EUR/MWh]"] = (GBWeight2["Gas Factor [EUR/MWh]"]-GBWeight2["Gas Factor Unc. [EUR/MWh]"])
GBWeight2["Gas Factor Max. [EUR/MWh]"] = (GBWeight2["Gas Factor [EUR/MWh]"]+GBWeight2["Gas Factor Unc. [EUR/MWh]"])
GBWeight2["EP Gas-Corrected"] = GBWeight["Electricity Price [EUR/MWh]"] - GBWeight2["Gas Factor [EUR/MWh]"]

# Add the data for wind based on the correlation coefficient
GBWeight2["Wind Factor [EUR/MWh]"] = MGBW*GBWeight["Wind Power [MWh]"]+CGBW
GBWeight2["Wind Factor Unc. [EUR/MWh]"] = abs((1+rhoGBW_BZNGB)*(MGBW*GBWeight["Wind Power [MWh]"]+CGBW))
GBWeight2["Wind Factor Min. [EUR/MWh]"] = (GBWeight2["Wind Factor [EUR/MWh]"]-GBWeight2["Wind Factor Unc. [EUR/MWh]"])
GBWeight2["Wind Factor Max. [EUR/MWh]"] = (GBWeight2["Wind Factor [EUR/MWh]"]+GBWeight2["Wind Factor Unc. [EUR/MWh]"])
```

```python
# Plot gas and wind based predictions without weighting but with correlation for uncertainty
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

plt.title('Predicting Day-Ahead Electricity Price of Great Britain with Correlation-based Uncertainty')
plt.ylabel('Day-Ahead Price of Electricity DAEP [EUR/MWh]')

MaxGas = GBWeight2["Gas Factor Max. [EUR/MWh]"].rolling(28,  center=True).mean()
MinGas = GBWeight2["Gas Factor Min. [EUR/MWh]"].rolling(28,  center=True).mean()

MaxWin = GBWeight2["Wind Factor Max. [EUR/MWh]"].rolling(28,  center=True).mean()
MinWin = GBWeight2["Wind Factor Min. [EUR/MWh]"].rolling(28,  center=True).mean()

plt.fill_between(GBWeight2["Date"], MinWin, MaxWin, color="lightblue")
plt.fill_between(GBWeight2["Date"], MinGas, MaxGas, color="lightyellow")
ax.plot(GBWeight2["Date"], GBWeight2["Electricity Price [EUR/MWh]"], color = 'green', label="DAEP Actual")
ax.plot(GBWeight2["Date"], GBWeight2["Gas Factor [EUR/MWh]"].rolling(28, center=True).mean(), linewidth=3, color='gold', label="Predicted DAEP from SAP")
ax.plot(GBWeight2["Date"], GBWeight2["Wind Factor [EUR/MWh]"].rolling(28, center=True).mean(), linewidth=3, color='steelblue', label="Predicted DAEP from Wind")
ax.plot(GBWeight2["Date"], GBWeight2["Electricity Price [EUR/MWh]"].rolling(28,  center=True).mean(), linewidth=3, color = "darkred", label="DAEP Rolling Average")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

fig.autofmt_xdate()

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.xlim(18628, 19358)
plt.ylim(0, 700)
plt.tight_layout()

ax.xaxis.set_minor_locator(mdates.MonthLocator())
ax.xaxis.set_minor_formatter(mdates.DateFormatter(""))

plt.show()
```
