```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

BZFRDA2021 = pd.read_excel(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\BZ_FR_20210101-20211231.xlsx")

BZFRDA2021data = pd.DataFrame({"BZN|FR 2021": BZFRDA2021.iloc[:,1]})
BZFRDA2021data["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")

NewOrder = ["Date", "BZN|FR 2021"]

#BZFRDAdata["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")
BZFRDA2021data = BZFRDA2021data.reindex(columns=NewOrder)
BZFRDA2021data
```

```python
BZFRDA2022a = pd.read_excel(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\BZ_FR_20220101-20220331.xlsx")

BZFRDA2022dataa = pd.DataFrame({"BZN|FR 20220101-20220331": BZFRDA2022a.iloc[:,1]})
BZFRDA2022dataa["Date"] = pd.date_range(start="1/1/2022", end="3/31/2022", freq="D")

NewOrder = ["Date", "BZN|FR 20220101-20220331"]

BZFRDA2022dataa = BZFRDA2022dataa.reindex(columns=NewOrder)
BZFRDA2022dataa
```

```python
BZFRDA2022b = pd.read_csv(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\BZ_FR_20220401-20230131.csv")

BZFRDA2022b = BZFRDA2022b.replace('-', np.NaN)

print(BZFRDA2022b)

BZFRDA2022datab = pd.concat([BZFRDA2022b['Apr-22'], BZFRDA2022b['May-22'], BZFRDA2022b['Jun-22'], BZFRDA2022b['Jul-22'], BZFRDA2022b['Aug-22'], BZFRDA2022b['Sep-22'], BZFRDA2022b['Oct-22'], BZFRDA2022b['Nov-22'], BZFRDA2022b['Dec-22']], ignore_index=True).to_frame("BZN|FR 20220401-20221231")
BZFRDA2022datab = BZFRDA2022datab.dropna()
BZFRDA2022datab["Date"] = pd.date_range(start="4/1/2022", end="12/31/2022", freq="D")

BZFRDA2022datab["BZN|FR 20220401-20221231"] = BZFRDA2022datab["BZN|FR 20220401-20221231"].replace(to_replace=",", value=".", regex=True)

NewOrder = ["Date", "BZN|FR 20220401-20221231"]

#BZFRDAdata["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")
BZFRDA2022datab = BZFRDA2022datab.reindex(columns=NewOrder)

BZFRDA2022datab.reset_index(drop=True, inplace=True)
BZFRDA2022datab
```

```python
BZFRDA = pd.concat([BZFRDA2021data["BZN|FR 2021"], BZFRDA2022dataa["BZN|FR 20220101-20220331"], BZFRDA2022datab["BZN|FR 20220401-20221231"]], ignore_index=True).to_frame("BZN|FR")

BZFRDA["Date"] = pd.date_range(start="1/1/2021", periods=730, freq="D")

NewOrder = ["Date", "BZN|FR"]

BZFRDA["BZN|FR"] = pd.to_numeric(BZFRDA["BZN|FR"], errors='coerce')
BZFRDA = pd.DataFrame(BZFRDA)

BZFRDA["BZN|FR"] = BZFRDA["BZN|FR"].round(2)

BZFRDA = BZFRDA.reindex(columns=NewOrder)

BZFRDA
```

```python
from matplotlib.ticker import MultipleLocator
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

plt.title('FR Electricity Day-Ahead Price')
plt.ylabel('Day-Ahead Electricty Price [EUR/MWh]')

ax.plot(BZFRDA["Date"], BZFRDA["BZN|FR"], color = "violet", label="FR DA")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

fig.autofmt_xdate()

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.xlim(18628, 19358)
plt.ylim(0, 800)
plt.tight_layout()

plt.show()
print(ax.get_xlim())
```
