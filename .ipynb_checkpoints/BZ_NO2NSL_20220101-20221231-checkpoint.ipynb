```python
# Python code to find average of each consecutive segment
  
# Importing
import statistics
from itertools import islice
import numpy as np

# Change CET-CEST 25/26-Mar-21 i.e. row 2039-2040
# Change CEST-CET 29/30-Oct-21 i.e. row 7247-7248

# 01-Oct-21 00:00-01:00 is in column 6552

# Days til CET-CEST change from 01-Jan-21
dtcest = int(len(BZNO2NSLDA2022.iloc[0:(2039+1),0])/24)
print(dtcest)

#Days til CEST-CET change from 01-Jan-21
dtcst = int(len(BZNO2NSLDA2022.iloc[0:(7247+1),0])/24)
print(dtcst)

def function():
    result = []
    for i in range(0, len(BZNO2NSLDA2022.iloc[:,0])):
        if i < dtcest:
            number = i
            start = 24*i
            end = 24*(i+1)-1+1
            day = BZNO2NSLDA2022.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2022.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif i == 85: # For some reason this day does not show in the DataFrame
            number = i
            start = 24*i
            skip = 2042
            end = 2063+1
            a = sum(BZNO2NSLDA2022.iloc[start:skip,1])
            b = sum(BZNO2NSLDA2022.iloc[skip+1:end,1])
            day = BZNO2NSLDA2022.iloc[start,0]
            operation = (a+b)/(len(BZNO2NSLDA2022.iloc[start:end,1])-1)
            result.append({"Day": day, "Daily DA": operation})
        elif 86 <= i < dtcst:
            number = i
            start = 24*i
            end = 24*(i+1)-1+1
            day = BZNO2NSLDA2022.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2022.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif i == 302:
            number = i
            start = 24*i
            end = 7272+1
            day = BZNO2NSLDA2022.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2022.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})
        elif 303 <= i < 365:
            number = i
            start = 24*i+1
            end = 24*(i+1)-1+2
            day = BZNO2NSLDA2022.iloc[start,0]
            operation = statistics.mean(BZNO2NSLDA2022.iloc[start:end,1])
            result.append({"Day": day, "Daily DA": operation})   
    df = pd.DataFrame(result, columns=['Day','Daily DA'])
    return df
print(function())
```
