```python
import matplotlib.pyplot as plt

# Set x-axis
Quarter = UKEGS.iloc[43,2:]

# Set y-axis data series
Coal = UKEGS.iloc[69,2:]
Oil = UKEGS.iloc[70,2:]
Gas = UKEGS.iloc[71,2:]

# Create a subplot of given dimensions
fig, ax = plt.subplots()
fig.set_figheight(4)
fig.set_figwidth(10)

plt.title('Quarterly High-Carbon Intensity Electricity Generation in Great Britain (GB)')
plt.ylabel('Electricity Generated [TWh]')

# Plot individual data series one by one
ax.plot(Quarter, Gas, label="Gas")
ax.plot(Quarter, Coal, label="Coal")
ax.plot(Quarter, Oil, label="Oil")

plt.legend()

# Set gridline styles
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

# Set the x-axis tick interval and add minor ticks on the y-axis
n = 4
ax.xaxis.set_minor_locator(plt.MultipleLocator(n-3))
ax.xaxis.set_major_locator(plt.MultipleLocator(n))
ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

print(ax.get_xlim())
print(ax.get_ylim())

# Format the labels on the x-axis
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.xlim(0, 98)
plt.ylim(0, 55)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
# Set x-axis
x = UKEGS.iloc[43,2:]

# corresponding y axis values
Nuclear = UKEGS.iloc[72,2:]
NFH = UKEGS.iloc[73,2:]
TWind = UKEGS.iloc[74,2:]
Tidal = UKEGS.iloc[77,2:]
Solar = UKEGS.iloc[78,2:]
Bioenergy = UKEGS.iloc[79,2:]
PSH = UKEGS.iloc[81,2:]

fig, ax = plt.subplots()
fig.set_figheight(4)
fig.set_figwidth(10)

plt.title('Quarterly Low-Carbon Intensity Electricity Generation in Great Britain (GB)')
plt.ylabel('Electricity Generated [TWh]')

ax.plot(x, TWind, label="Wind")
ax.plot(x, Nuclear, label="Nuclear")
ax.plot(x, Bioenergy, label="Bioenergy")
ax.plot(x, Solar, label="Solar")
ax.plot(x, NFH, label="NFH")
ax.plot(x, PSH, label="PSH")
ax.plot(x, Tidal, label="Tidal")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

n = 4
ax.xaxis.set_minor_locator(plt.MultipleLocator(n-3))
ax.xaxis.set_major_locator(plt.MultipleLocator(n))
ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

print(ax.get_xlim())
print(ax.get_ylim())

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.xlim(0, 98)
plt.ylim(0, 30)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
# Set x-axis
x = UKEGS.iloc[43,2:]

# corresponding y axis values
TWind = UKEGS.iloc[74,2:]
OnWind = UKEGS.iloc[75,2:]
OffWind = UKEGS.iloc[76,2:]

fig, ax = plt.subplots()
fig.set_figheight(4)
fig.set_figwidth(10)

plt.title('Quarterly Wind Electricity Generation in Great Britain (GB)')
plt.ylabel('Electricity Generated [TWh]')

ax.plot(x, TWind, label="Total Wind")
ax.plot(x, OffWind, label="Offshore Wind")
ax.plot(x, OnWind, label="Onshore Wind")

plt.legend()
plt.grid(visible=True, which='major', color='0.65', linestyle='-')
plt.grid(visible=True, which='minor', color='0.65', linestyle=':')

n = 4
ax.xaxis.set_minor_locator(plt.MultipleLocator(n-3))
ax.xaxis.set_major_locator(plt.MultipleLocator(n))
ax.yaxis.get_ticklocs(minor=True)
ax.minorticks_on()

print(ax.get_xlim())
print(ax.get_ylim())

plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=45, ha='right')
plt.xlim(0, 98)
plt.ylim(0, 30)
plt.tight_layout()

plt.show()
```
