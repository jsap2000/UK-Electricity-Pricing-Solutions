```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(7,8), sharex=True)

fig.suptitle("Energy Mix in the UK and its Neighbours")

FuMi = ["Bioenergy", "Coal", "Gas", "Hydro", "Nuclear", "Other Fossil", "Other Renewables", "Solar", "Wind"]
UKAnGeTWh = UKEMix["Annual Generation [%]"]
NOAnGeTWh = NOEMix["Annual Generation [%]"]
NLAnGeTWh = NLEMix["Annual Generation [%]"]
BEAnGeTWh = BEEMix["Annual Generation [%]"]
FRAnGeTWh = FREMix["Annual Generation [%]"]
IEAnGeTWh = IEEMix["Annual Generation [%]"]

ax1.barh(FuMi, UKAnGeTWh, color = "navy")
ax2.barh(FuMi, NOAnGeTWh, color = "firebrick")
ax3.barh(FuMi, NLAnGeTWh, color = "darkorange")
ax4.barh(FuMi, BEAnGeTWh, color = "black")
ax5.barh(FuMi, FRAnGeTWh, color = "orchid")
ax6.barh(FuMi, IEAnGeTWh, color = "mediumseagreen")

ax1.set_title("United Kingdom")
ax2.set_title("Norway")
ax3.set_title("The Netherlands")
ax4.set_title("Belgium")
ax5.set_title("France")
ax6.set_title("Republic of Ireland")

ax5.set_xlabel("Electricity Generation Contribution [%]")
ax6.set_xlabel("Electricity Generation Contribution [%]")
    
for ax in fig.get_axes():
    ax.label_outer()
    ax.grid(visible=True, which='major', color='0.65', linestyle='-')
    ax.grid(visible=True, which='minor', color='0.65', linestyle=':')
    
ax.set_xlim(0,100)

plt.tight_layout()
plt.show()
```
