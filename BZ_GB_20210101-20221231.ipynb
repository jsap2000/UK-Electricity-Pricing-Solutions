This is where I imported the data for 2022. I had to stack these cells together. This is the data I captured on the Nord Pool ASA website.

```python
BZGBDA2022 = pd.read_csv(r"C:\Users\anton_zpf58tt\OneDrive\Documents\Glasgow\Year 4\ENG4110P Individual Project 4\Code\UK-Electricity-Pricing-Solutions\BZ_GB_20220101-20221231.csv")

BZGBDA2022 = BZGBDA2022.replace('-', np.NaN)

print(BZGBDA2022)
#BZGBDA2021data = pd.DataFrame({"BZN|GB": BZGBDA2021.iloc[0:365,2]})
#BZGBDA2021data["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")

#NewOrder = ["Date", "BZN|GB"]

#BZGBDAdata["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")
#BZGBDA2021data = BZGBDA2021data.reindex(columns=NewOrder)
#BZGBDA2021data

#BZGBDA2022data = pd.DataFrame({"BZN|GB": BZGBDA2022.iloc[0:,4] BZGBDA2022.iloc[0:,3]})
#BZGBDA2022data

#BZGBDA2022data = pd.DataFrame(BZGBDA2022.stack()) #.reset_index()
#BZGBDA2022data = BZGBDA2022.stack(level = ["Jan-22", "Feb-22"])

BZGBDA2022data = pd.concat([BZGBDA2022['Jan-22'], BZGBDA2022['Feb-22'], BZGBDA2022['Mar-22'], BZGBDA2022['Apr-22'], BZGBDA2022['May-22'], BZGBDA2022['Jun-22'], BZGBDA2022['Jul-22'], BZGBDA2022['Aug-22'], BZGBDA2022['Sep-22'], BZGBDA2022['Oct-22'], BZGBDA2022['Nov-22'], BZGBDA2022['Dec-22']], ignore_index=True).to_frame("BZN|GB 2022")
BZGBDA2022data = BZGBDA2022data.dropna()
BZGBDA2022data["Date"] = pd.date_range(start="1/1/2022", periods=365, freq="D")

BZGBDA2022data["BZN|GB 2022"] = BZGBDA2022data["BZN|GB 2022"].replace(to_replace=",", value=".", regex=True)

NewOrder = ["Date", "BZN|GB 2022"]

#BZGBDAdata["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")
BZGBDA2022data = BZGBDA2022data.reindex(columns=NewOrder)

BZGBDA2022data.reset_index(drop=True, inplace=True)
BZGBDA2022data
```

Here I simply stacked the data from 2021 together with that from 2022.

```python
BZGBDA = pd.concat([BZGBDA2021data["BZN|GB 2021"], BZGBDA2022data["BZN|GB 2022"]], ignore_index=True).to_frame("BZN|GB")

BZGBDA["Date"] = pd.date_range(start="1/1/2021", periods=730, freq="D")

NewOrder = ["Date", "BZN|GB"]

#BZGBDAdata["Date"] = pd.date_range(start="1/1/2021", periods=365, freq="D")
BZGBDA = BZGBDA.reindex(columns=NewOrder)

BZGBDA
```
