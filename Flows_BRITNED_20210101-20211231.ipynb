```python
# Python code to find average of each consecutive segment
  
# Importing
import statistics
from itertools import islice
import numpy as np

# Change CET-CEST 27/28-Mar-21 i.e. row 2063-2064
# Change CEST-CET 30/31-Oct-21 i.e. row 7271-7272

# 01-Oct-21 00:00-01:00 is in column 6552

# Days til CET-CEST change from 01-Jan-21
dtcest = int(len(BND2021.iloc[0:(2063+1),0])/24)
print(dtcest)

#Days til CEST-CET change from 01-Jan-21
dtcst = int(len(BND2021.iloc[0:(7271+1),0])/24)
print(dtcst)

def function():
    result = []
    for i in range(0, len(BND2021.iloc[:,0])):
        if i < dtcest:
            number = i
            start = 24*i
            end = 24*(i+1)-1+1
            day = BND2021.iloc[start,0]
            operation1 = sum(BND2021.iloc[start:end,1])
            operation2 = sum(BND2021.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif i == 86: # For some reason this day does not show in the DataFrame
            number = i
            start = 24*i
            skip = 2066
            end = 2087+1
            a = sum(BND2021.iloc[start:skip,1])
            b = sum(BND2021.iloc[skip+1:end,1])
            day = BND2021.iloc[start,0]
            operation1 = (a+b)/(len(BND2021.iloc[start:end,1])-1)
            operation2 = (a+b)/(len(BND2021.iloc[start:end,2])-1)
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif 87 <= i < dtcst:
            number = i
            start = 24*i
            end = 24*(i+1)-1+1
            day = BND2021.iloc[start,0]
            operation1 = sum(BND2021.iloc[start:end,1])
            operation2 = sum(BND2021.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif i == 303:
            number = i
            start = 24*i
            skip = 7275
            end = 7296+1
            day = BND2021.iloc[start,0]
            operation1 = sum(BND2021.iloc[start:skip,1]) + sum(BND2021.iloc[(skip+1):end,1])
            operation2 = sum(BND2021.iloc[start:skip,2]) + sum(BND2021.iloc[(skip+1):end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif 304 <= i < 365:
            number = i
            start = 24*i+1
            end = 24*(i+1)-1+2
            day = BND2021.iloc[start,0]
            operation1 = sum(BND2021.iloc[start:end,1])
            operation2 = sum(BND2021.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})   
    df = pd.DataFrame(result, columns=['Day','BZN|NL > BZN|GB [MW]', 'BZN|GB > BZN|NL [MW]'])
    return df
print(function())
```
