```python
# Python code to find average of each consecutive segment
  
# Importing
import statistics
from itertools import islice
import numpy as np

def function():
    result = []
    for i in range(0, len(BND2022.iloc[:,0])):
        if i == 0:
            start = i
            end = 92+1
            day = BND2022.iloc[start,0]
            operation1 = sum(BND2022.iloc[start:end,1])
            operation2 = sum(BND2022.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif 0 < i < 85:
            start = 93+(i-1)*96
            end = 189+(i-1)*96
            day = BND2022.iloc[start,0]
            operation1 = sum(BND2022.iloc[start:end,1])
            operation2 = sum(BND2022.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif i == 85: # For some reason this day does not show in the DataFrame
            start = 93+(i-1)*96
            skip = 8165
            end = 8252+1
            day = BND2022.iloc[start,0]
            operation1 = sum(BND2022.iloc[start:skip,1]) + sum(BND2022.iloc[(skip+4):end,1])
            operation2 = sum(BND2022.iloc[start:skip,2]) + sum(BND2022.iloc[(skip+4):end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif 86 <= i < 302:
            start = 96*i-3
            end = 96*(i+1)-3
            day = BND2022.iloc[start,0]
            operation1 = sum(BND2022.iloc[start:end,1])
            operation2 = sum(BND2022.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif i == 302:
            start = 96*i-3
            end = 29088+1
            day = BND2022.iloc[start,0]
            operation1 = sum(BND2022.iloc[start:end,1])
            operation2 = sum(BND2022.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})
        elif 303 <= i < 365:
            start = 96*i+1
            end = 96*(i+1)+1
            day = BND2022.iloc[start,0]
            operation1 = sum(BND2022.iloc[start:end,1])
            operation2 = sum(BND2022.iloc[start:end,2])
            result.append({"Day": day, "BZN|NL > BZN|GB [MW]": operation1, "BZN|GB > BZN|NL [MW]": operation2})   
    df = pd.DataFrame(result, columns=['Day','BZN|NL > BZN|GB [MW]', 'BZN|GB > BZN|NL [MW]'])
    return df
print(function())
```
